```python
#path es la ruta de acceso: absoluta(desde raiz), relativa(desde dir actual)
def leerArch(path):
  modo='r'
  with open(path, modo) as f:
    data = f.read()
  return data
  
path='files/data.txt' #relativa
data = leerArch(path)
print(data)
```

```python
#4 operaciones sobre un archivo: crear, abrir, cerrar y eliminar
#path relativo/absoluto
#Absoluto, doy todo el path desde raiz
#Relativo, a partir de la carpeta donde se ejecuta el codigo
#instruccion basica with open(path, modo) as ff: 
#open abre el archivo, en el modo dado. 
#with cierra automaticamente el archivo despues de su uso
#Modos o permisos: 'r': lectura, 'w':escritura, 'a': append
#w: crea el arch si no existe o lo sobrescribe si ya existe
#a: agrega contenido al final, o lo crea si no existe
```

```python
#Ejemplo de escritura
path="files/wdata.txt"
data = "Estamos escribiendo en el archivo wdata.txt..."
modo='w'#el apuntador se ubica al principio del arch
with open(path, modo) as f:
  f.write(data +'\n')
  f.write(data +'\tabc\n')
  f.write(data +'\tpqr\n')
#comprobacion
print(leerArch(path))
```

```python
#Uso de append, agregar contenido nuevo al final del arch
path="files/wdata.txt"
data = "mas cosas ... "
modo='a'
print(leerArch(path))
#añadir
with open(path, modo) as f:
  f.write(data +'\tuno\n')
  f.write(data +'\tdos\n')
  f.write(data +'\ttres\n')
#comprobacion
print(leerArch(path))
```

```python
#Otros modos o permisos 'r+': lectura /escritura, puntero al inicio del arch
#'a+': añadir/leer, leer/escribir, el puntero al final del arch
#'-x': exclusivamente usado para crear un arch, fallara si ya existe el arch
#'b': se añade al final del modo para tratar con arch binarios
#ejemplo: 'rb', 'wb', 'ab', 'r+b', 'a+b'
#lectura, puede ser un arch de txt , la salida quedara como un string
#si el arch es binario, la salida quedara como objetos en bytes
#Parametro encoding='utf-8' si el arch viene con el juego de caracters especifico
path="files/data1.txt"
with open(path, "r",encoding='utf-8') as f:
  print(f.read())
```

```python
#leer byte a byte, n: un parametro entero que especifica cuantos bytes se leen y 
#deja el apuntador justo despues del ultimo byte leido.
path="files/data1.txt"
n=5
with open(path, "r") as f:
  linea1 = f.read(n+2)
  linea2 = f.read(n+7)
print(linea1)
print(linea2)
#comprobacion
print(leerArch(path))
```

```python
#readline(), leer un archivo linea x linea, el parametro encoding para interpretar vocales con tildes
path="files/data1.txt"
with open(path, "r", encoding="utf-8") as f:
  linea1 = f.readline() # fin de linea:  \n
  linea2 = f.readline()
print(linea1, end="")
print(linea2, end="")
#comprobacion
print(leerArch(path))
```

```python
#Para escribir en archivos planos, datos que no son cadenas, 
#Deberia convertir cada dato que no sea string a string str()
data = [1234, 5678, 9012, 3.14159265, False]
path="files/data3.txt"
with open(path, "w+") as f:
  for valor in data:
    str_val = str(valor)
    f.write(str_val)
    f.write("\n")
#comprobacion
print(leerArch(path))
```

```python
#seek(offset, from_what), permite ubicar el apuntador  en alguna parte del archivo 
#mueve el apuntador al principio, al final o a alguna posic especifica de un archivo
offset=11; #especifica nro de caracteres de la posicion base
base=0 # 0: desde el principio; 1: posicion actual; 2: al final del arch
path="files/data4.txt"
print(leerArch(path))
#f=0
#print(f)
with open(path, "r", encoding="utf-8") as f:
  f.seek(offset, base) #reposiciona el apuntador 
  for line in f:
    print(line, end="")
```

```python
#Metodo tell() Determina el tamaño de un archivo en bytes
path='files/data4.txt'
with open(path, "a") as f:
  print('Tamaño del arch: ', path,'=', f.tell())
```

```python
#edicion de archivos
# Abre el archivo en modo de solo lectura y lo guarda en una lista
#Luego Utilizar el metodo list.insert(i, x) que inserta nuevos datos a la lista
#Luego unir con join() y reescribir todo al archivo
path="files/data5.txt"
print(leerArch(path))
with open(path, "r", encoding="utf-8") as f:
  list_content = f.readlines() #El contenido queda en una lista
list_content.insert(1, "Esta es la linea 1.5: jejeje...\n")#modifica la lista
# Re-abre el archivo en modo de solo escritura
# para sobreescribir la version anterior 

with open(path, "w", encoding="utf-8") as f:
  contenido = "".join(list_content)
  f.write(contenido)
#Comprobacion
print(leerArch(path))
```

```python
#metodo scandir(), Lista archivos de una carpeta 
#import os: este modulo trae funciones de utilidad para remover archs (remove())
#crear directorios mkdid(path), retornar name, is_dir(), tamaño: stat().st_size(), etc
import os
path="files/"
entries = os.scandir(path)
for entry in entries:
  print(entry.name + ", es dir: " + str(entry.is_dir()) 
  + "\t, size: " + str(entry.stat().st_size) + " bytes.")
```

```python
#Guardar estructuras de datos (diccionarios o listas) en archivos, 
#metodo dump() de libreria pickle serializa los datos
import pickle 
#In Python 3, zip returns an iterator of tuples 
#To get a list of tuples, use list(zip(a, b, ...))
name = ["mohit", "bhaskar", "manish"]
skill = ["Python", "Python", "Java"]

dict1 = dict([(k,v) for k, v in zip(name, skill)])
path="files/programming_powers.pkl"
with open(path, "wb") as p_file:
  pickle.dump(name, p_file)
  pickle.dump(skill, p_file)
  pickle.dump(dict1, p_file)
#comprobacion en la siguiente celda
```

```python
# metodo load(), Carga estructuras de datos desde archivos, inv a serializac
import pickle
path="files/programming_powers.pkl"
with open(path, "rb") as p_file:
  list1 = pickle.load(p_file)
  list2 = pickle.load(p_file)
  dict1 = pickle.load(p_file)
print(list1)
print(list2)
print(dict1)
```

```python
#P1- Dado el archivo de texto files/SalesJan2009.csv, procese el archivo
#para obtener el nro de compras realizadas en un pais dado. Ejemplo:
path="files/SalesJan2009.csv"
atributo='Country'
vrAtri = input('Ingrese '+ atributo +': ') # United Kingdom
with open(path, "r", encoding="utf-8") as f: 
  cabecera = f.readline()
  llaves = cabecera.split(',') #cabeza es una cadena
  pos = llaves.index(atributo)
  n=0 #contador
  for linea in f: #lee linea x linea
    vals = linea.split(',') #linea es una cadena
    if vrAtri == vals[pos]: #vals es una lista (comparo cadenas)
      n+=1 #contador
print(n)
print(llaves)
```

```python
#P2-Dado el archivo de texto files/SalesJan2009.csv, procese el archivo
#para obtener el nro de compras realizadas con un medio de pago dado.
path="files/SalesJan2009.csv"
with open(path, "r", encoding="utf-8") as f: 
  cabecera = f.readline() #retorna una cadena
  campos=cabecera.split(',')#retorna una lista
  i = list(range(len(campos)))#coleccion 0, 1, 2, 3, ... len()
  dict1 = dict([(k,v) for k, v in zip(i, campos)])
  print(dict1)
  pos = int(input('ingrese llave del campo deseado: '))# 3
  atributo = str(dict1.get(pos)) #atributo='Payment_Type'
  vrAtri = input('Ingrese '+ atributo +': ') #Visa
  n=0
  for linea in f: #linea va tomando una linea de f a la vez
    val = linea.split(',') #linea, val es lista de cadenas
    if vrAtri == val[pos]: #Comparacion de strings
      n+=1
#ya se cerro el arch
print('Nro de compras con ', vrAtri,'= ',n)
```
